```python
import	numpy	as	np
import	scipy
from scipy import integrate
import	sympy
import	cv2
import	matplotlib as mpl
import	matplotlib.pyplot	as	plt
from	math import *
```

Начальные условия:

```python
t = np.linspace(0, 11, 500)
k0 = 0.02
x0 = 10					#начальное кол-во команды 1
#x0 = np.linspace(10, 0, 50)
y0 = [2, 4, 6, 8, 10]	#начальное кол-во команды 2
#y0 = np.linspace(2, 0, 50)
k1 = [0.1, 0.2, 0.05]	#{1/x0, 2/x0, 1/2*x0}
```

Функция Риккати

```python
def Vx(s, n, k, t):
    dxdt = - k0 * (n+dxdt * t) * (s+dydt*t) * exp( -k * ((n+dxdt * t) - (s+dydt*t)))
    return dxdt

def Vy(s, n, k, t):
    dydt = - k0 * (n+dxdt * t) * (s+dydt*t) * exp(  k * ((n+dxdt * t) - (s+dydt*t)))
    return dydt
```

```python
def V(s, n, k, t):
    dxdt = - k0 * (n+dxdt * t) * (s+dydt*t) * exp( -k * ((n+dxdt * t) - (s+dydt*t)))
    dydt = - k0 * (n+dxdt * t) * (s+dydt*t) * exp(  k * ((n+dxdt * t) - (s+dydt*t)))
    return [dxdt,dydt]
```

```python
i=0
j=0
wx =x0 + Vx(y0[0], x0, k1[0], t)
wy =y0[0] + Vy(y0[0], x0, k1[0], t)
w = V(y0[0], x0, k1[0], t)
#print(wx,"\n", wy)
print(w)
```
